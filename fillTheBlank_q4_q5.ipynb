# **Q5 Fill in**

```python
import cv2
import easyocr
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
import torch
from difflib import SequenceMatcher

# --- Step 1: Load Image ---
image_path = "/content/iogtrmot4kot4rio.jpeg"  # Replace with your image
image = cv2.imread(image_path)

# --- Step 2: Detect text with EasyOCR ---
reader = easyocr.Reader(['en'])
results = reader.readtext(image, detail=1)

# --- Step 3: Find Q4 Box ---
q4_box = None
for (bbox, text, conf) in results:
    if "4" in text.strip() or "Q4" in text.strip():
        q4_box = bbox
        print("üìå Found Q4 at:", bbox, "with text:", text)
        break

if not q4_box:
    raise ValueError("‚ùå Could not find Q4 in the image.")

# --- Step 4: Crop Handwritten Answer ---
x_min = min(pt[0] for pt in q4_box)
x_max = max(pt[0] for pt in q4_box)
y_top = min(pt[1] for pt in q4_box) - 10
y_bottom = max(pt[1] for pt in q4_box) + 10
x_left = max(pt[0] for pt in q4_box) + 5  # slightly right of the printed question
x_right = x_left + 100  # narrower width to isolate answer

handwritten_crop = image[y_top:y_bottom, x_left:x_right]

plt.imshow(cv2.cvtColor(handwritten_crop, cv2.COLOR_BGR2RGB))
plt.title("‚úçÔ∏è Handwritten Answer Area")
plt.axis("off")
plt.show()

# --- Step 5: Clean Preprocessing ---
gray = cv2.cvtColor(handwritten_crop, cv2.COLOR_BGR2GRAY)
resized = cv2.resize(gray, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
rgb = cv2.cvtColor(resized, cv2.COLOR_GRAY2RGB)
pil_image = Image.fromarray(rgb)

plt.imshow(pil_image)
plt.title("‚ú® Cleaned Input for TrOCR")
plt.axis("off")
plt.show()

# --- Step 6: TrOCR Inference ---
processor = TrOCRProcessor.from_pretrained("microsoft/trocr-base-handwritten")
model = VisionEncoderDecoderModel.from_pretrained("microsoft/trocr-base-handwritten")

pixel_values = processor(images=pil_image, return_tensors="pt").pixel_values
generated_ids = model.generate(pixel_values)
recognized_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

print("üìò Recognized Handwritten Answer:", recognized_text)

# --- Step 7: Fuzzy Token Grading ---
recognized_clean = recognized_text.lower().strip()
tokens = recognized_clean.split()
correct_answer = "relu"
threshold = 0.5

if tokens:
    best_match = max(tokens, key=lambda t: SequenceMatcher(None, t, correct_answer).ratio())
    score = SequenceMatcher(None, best_match, correct_answer).ratio()

    print(f"üîç Best token match: '{best_match}' with {round(score * 100, 2)}% similarity")

    if score >= threshold:
        print("‚úÖ Grade: Correct (1/1)")
    else:
        print(f"‚ùå Grade: Incorrect (0/1) ‚Äî Expected: {correct_answer}")
else:
    print("‚ö†Ô∏è No tokens found in recognized text.")

# --- Step 2: Detect text with EasyOCR ---
reader = easyocr.Reader(['en'])
results = reader.readtext(image, detail=1)

# --- Step 3: Find Q5 Box ---
q5_box = None
for (bbox, text, conf) in results:
    if "5" in text.strip() or "Q5" in text.strip():
        q5_box = bbox
        print("üìå Found Q5 at:", bbox, "with text:", text)
        break

if not q5_box:
    raise ValueError("‚ùå Could not find Q5 in the image.")

# --- Step 4: Crop Handwritten Answer ---
x_min = min(pt[0] for pt in q5_box)
x_max = max(pt[0] for pt in q5_box)
y_top = min(pt[1] for pt in q5_box) - 5
y_bottom = max(pt[1] for pt in q5_box) + 5
x_left = max(pt[0] for pt in q5_box) + 30 # slightly right of the printed question
x_right = x_left + 200  # narrower width to isolate answer

handwritten_crop = image[y_top:y_bottom, x_left:x_right]

plt.imshow(cv2.cvtColor(handwritten_crop, cv2.COLOR_BGR2RGB))
plt.title("‚úçÔ∏è Handwritten Answer Area")
plt.axis("off")
plt.show()

# --- Step 5: Clean Preprocessing ---
gray = cv2.cvtColor(handwritten_crop, cv2.COLOR_BGR2GRAY)
resized = cv2.resize(gray, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
rgb = cv2.cvtColor(resized, cv2.COLOR_GRAY2RGB)
pil_image = Image.fromarray(rgb)

plt.imshow(pil_image)
plt.title("‚ú® Cleaned Input for TrOCR")
plt.axis("off")
plt.show()

# --- Step 6: TrOCR Inference ---
processor = TrOCRProcessor.from_pretrained("microsoft/trocr-base-handwritten")
model = VisionEncoderDecoderModel.from_pretrained("microsoft/trocr-base-handwritten")

pixel_values = processor(images=pil_image, return_tensors="pt").pixel_values
generated_ids = model.generate(pixel_values)
recognized_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

print("üìò Recognized Handwritten Answer:", recognized_text)

# --- Step 7: Fuzzy Token Grading ---
recognized_clean = recognized_text.lower().strip()
tokens = recognized_clean.split()
correct_answer = "lasso"
threshold = 0.5

if tokens:
    best_match = max(tokens, key=lambda t: SequenceMatcher(None, t, correct_answer).ratio())
    score = SequenceMatcher(None, best_match, correct_answer).ratio()

    print(f"üîç Best token match: '{best_match}' with {round(score * 100, 2)}% similarity")

    if score >= threshold:
        print("‚úÖ Grade: Correct (1/1)")
    else:
        print(f"‚ùå Grade: Incorrect (0/1) ‚Äî Expected: {correct_answer}")
else:
    print("‚ö†Ô∏è No tokens found in recognized text.")
```
